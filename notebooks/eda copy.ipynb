```python
#outliers

def detect_outliers_iqr(dataframe, column):
    Q1 = dataframe[column].quantile(0.25)
    Q3 = dataframe[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = dataframe[
        (dataframe[column] < lower_bound) |
        (dataframe[column] > upper_bound)
    ]
    
    return outliers, lower_bound, upper_bound


for col in numeric_cols + ["total_budget", "roi"]:
    outliers, lower, upper = detect_outliers_iqr(df, col)
    print(f"\nVariable : {col}")
    print(f"Bornes : [{lower:.2f}, {upper:.2f}]")
    print(f"Nombre d'outliers : {len(outliers)}")
```

### Interprétation — détection des outliers

La méthode IQR détecte peu ou pas d’outliers sur `tv`, `sales` et `total_budget`. En revanche, `social_media` possède plusieurs valeurs atypiques et `roi` présente beaucoup d’outliers. Pour `social_media`, ces valeurs peuvent correspondre à des campagnes réellement plus orientées vers ce canal. Pour `roi`, les outliers doivent être interprétés avec prudence : un ROI élevé peut venir d’un petit budget total combiné à des ventes correctes. Il ne faut donc pas supprimer ces lignes automatiquement.


```python
#visualiser outliers

social_outliers, social_lower, social_upper = detect_outliers_iqr(df, "social_media")

print("Borne haute social_media :", social_upper)
display(
    social_outliers
    .sort_values(by="social_media", ascending=False)
    [["tv", "radio", "social_media", "influencer", "sales", "total_budget", "roi"]]
    .head(20)
)
```

# Modélisation prédictive

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler

from sklearn.dummy import DummyRegressor
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

## Prédiction des ventes

```python


target_sales = "sales"

features_sales = [
    "tv",
    "radio",
    "social_media"
]

X_sales = df[features_sales]
y_sales = df[target_sales]

X_train_sales, X_test_sales, y_train_sales, y_test_sales = train_test_split(
    X_sales,
    y_sales,
    test_size=0.2,
    random_state=42
)

print("Taille du jeu d'entraînement :", X_train_sales.shape)
print("Taille du jeu de test :", X_test_sales.shape)
```

Ce code prépare les données pour la première tâche de régression : la prédiction des ventes.

La variable `target_sales` correspond à la variable que l’on cherche à prédire, ici `sales`.

La variable `features_sales` contient les variables utilisées pour faire la prédiction :
- `tv` ;
- `radio` ;
- `social_media`.

Le dataset est ensuite séparé en deux parties :
- un jeu d’entraînement, utilisé pour apprendre au modèle ;
- un jeu de test, utilisé pour évaluer le modèle sur des données qu’il n’a jamais vues.

La séparation est faite avec `train_test_split`.

Le paramètre `test_size=0.2` 20 % des données sont gardées pour le test, et 80 % sont utilisées pour l’entraînement.


```python
# Fonction d'évaluation des modèles de régression

def evaluate_regression_model(model, X_train, X_test, y_train, y_test):
    """
    Entraîne un modèle de régression et calcule ses performances.
    """
    
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)
    
    return {
        "model": model,
        "y_pred": y_pred,
        "MAE": mae,
        "RMSE": rmse,
        "R2": r2
    }
```

```python
# Modèle baseline pour Sales

baseline_sales = DummyRegressor(strategy="mean")

baseline_sales_result = evaluate_regression_model(
    baseline_sales,
    X_train_sales,
    X_test_sales,
    y_train_sales,
    y_test_sales
)

baseline_sales_metrics = {
    "Model": "Baseline",
    "MAE": baseline_sales_result["MAE"],
    "RMSE": baseline_sales_result["RMSE"],
    "R2": baseline_sales_result["R2"]
}

baseline_sales_metrics
```

## Régression linéaire

```python
# Régression linéaire pour Sales

linear_sales_pipeline = Pipeline(steps=[
    ("scaler", StandardScaler()),
    ("model", LinearRegression())
])

linear_sales_result = evaluate_regression_model(
    linear_sales_pipeline,
    X_train_sales,
    X_test_sales,
    y_train_sales,
    y_test_sales
)

linear_sales_metrics = {
    "Model": "Linear Regression",
    "MAE": linear_sales_result["MAE"],
    "RMSE": linear_sales_result["RMSE"],
    "R2": linear_sales_result["R2"]
}

linear_sales_metrics
```

```python
# Comparaison Baseline vs Régression linéaire

sales_results_df = pd.DataFrame([
    baseline_sales_metrics,
    linear_sales_metrics
])

sales_results_df = sales_results_df.sort_values(by="RMSE")

display(sales_results_df)
```

## Interprétation des premiers modèles de prédiction des ventes

Les premiers résultats montrent une très forte différence entre le modèle baseline et la régression linéaire.

Le modèle baseline obtient :
- une MAE d’environ 80.78 ;
- une RMSE d’environ 93.18 ;
- un R² légèrement négatif.

Ce modèle prédit toujours la moyenne des ventes. Il ne prend donc pas en compte les budgets marketing. Ses performances sont faibles, ce qui est attendu pour un modèle de référence.

La régression linéaire obtient :
- une MAE d’environ 2.34 ;
- une RMSE d’environ 2.96 ;
- un R² d’environ 0.999.

Ces résultats indiquent que la régression linéaire prédit très bien les ventes à partir des budgets `tv`, `radio` et `social_media`.

Le R² très proche de 1 signifie que le modèle explique presque toute la variance des ventes. Autrement dit, les budgets marketing semblent fortement liés au volume de ventes.

La comparaison avec la baseline montre que la régression linéaire apporte une vraie valeur prédictive. Les erreurs moyennes sont beaucoup plus faibles que celles du modèle naïf.

Cependant, un score aussi élevé doit toujours être vérifié. Il faut s’assurer qu’il n’y a pas de fuite de données. Dans notre cas, la variable `roi` n’a pas été utilisée pour prédire `sales`, donc le modèle ne triche pas. Le très bon score s’explique probablement par une relation fortement linéaire entre les budgets marketing et les ventes.

```python
# Comparaison de plusieurs modèles pour Sales

models_sales = {
    "Baseline": DummyRegressor(strategy="mean"),
    "Linear Regression": Pipeline(steps=[
        ("scaler", StandardScaler()),
        ("model", LinearRegression())
    ]),
    "Ridge Regression": Pipeline(steps=[
        ("scaler", StandardScaler()),
        ("model", Ridge(alpha=1.0))
    ]),
    "Random Forest": RandomForestRegressor(
        n_estimators=200,
        random_state=42
    ),
    "Gradient Boosting": GradientBoostingRegressor(
        random_state=42
    )
}

sales_results = []
trained_sales_models = {}

for model_name, model in models_sales.items():
    result = evaluate_regression_model(
        model,
        X_train_sales,
        X_test_sales,
        y_train_sales,
        y_test_sales
    )
    
    trained_sales_models[model_name] = result["model"]
    
    sales_results.append({
        "Model": model_name,
        "MAE": result["MAE"],
        "RMSE": result["RMSE"],
        "R2": result["R2"]
    })

sales_results_df = pd.DataFrame(sales_results)
sales_results_df = sales_results_df.sort_values(by="RMSE")

display(sales_results_df)
```

## Visualisation des prédictions du meilleur modèle Sales

```python
# Visualisation des prédictions du meilleur modèle Sales

best_sales_model_name = sales_results_df.iloc[0]["Model"]
best_sales_model = trained_sales_models[best_sales_model_name]

y_pred_sales = best_sales_model.predict(X_test_sales)

plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_test_sales, y=y_pred_sales)

plt.plot(
    [y_test_sales.min(), y_test_sales.max()],
    [y_test_sales.min(), y_test_sales.max()],
    color="red",
    linestyle="--"
)

plt.title(f"Ventes réelles vs ventes prédites - {best_sales_model_name}")
plt.xlabel("Ventes réelles")
plt.ylabel("Ventes prédites")
plt.show()

print("Meilleur modèle Sales :", best_sales_model_name)
```
